## Step 4: Text Preprocessing Function
This function cleans any text before comparison

```python
# Creating the stemmer object
stemmer = PorterStemmer()

# Getting English stop words list from NLTK
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    """
    This function takes raw text and cleans it in 4 steps:
    Step 1 - Lowercase: converts all text to small letters
    Step 2 - Tokenize: splits sentence into individual words
    Step 3 - Remove stop words: removes useless common words
    Step 4 - Stem: reduces words to their root form
    """

    # Step 1: Convert to lowercase
    # Why: "Reset" and "reset" are same word, computer treats them differently
    text = text.lower()

    # Step 2: Tokenize - split into individual words
    # Why: algorithms work on words not full sentences
    tokens = word_tokenize(text)

    # Step 3: Remove stop words and non-alphabetic characters
    # Why: words like "the","is","how","do" appear everywhere and confuse matching
    tokens = [word for word in tokens if word.isalpha() and word not in stop_words]

    # Step 4: Stemming - reduce to root form
    # Why: "applying" and "apply" mean same thing, stemming makes them identical
    tokens = [stemmer.stem(word) for word in tokens]

    # Join back into a single string for TF-IDF
    return " ".join(tokens)

# Testing the function
sample = "How do I apply for revaluation of my exam papers?"
print("Original text:")
print(sample)
print("\nAfter preprocessing:")
print(preprocess_text(sample))
```

## Step 5: Building TF-IDF Vectors from FAQ Questions

```python
# Preprocess all FAQ questions
# We apply our cleaning function to every question in the dataset
preprocessed_questions = df['question'].apply(preprocess_text)

print("Preprocessed FAQ questions:")
print(preprocessed_questions.values)
```

```python
# Create TF-IDF Vectorizer
# This converts text into numerical vectors that can be mathematically compared

vectorizer = TfidfVectorizer()

# Fit and transform all FAQ questions into vectors
# fit - learns the vocabulary from your FAQ questions
# transform - converts each question into a numerical vector
faq_vectors = vectorizer.fit_transform(preprocessed_questions)

print(f"Shape of FAQ vectors: {faq_vectors.shape}")
print(f"This means: {faq_vectors.shape[0]} questions converted to {faq_vectors.shape[1]} dimensional vectors")
```

## Step 6: FAQ Matching Function Using Cosine Similarity

```python
def get_best_answer(user_question):
    """
    Takes user question and returns best matching FAQ answer
    using TF-IDF cosine similarity
    """

    # Handle empty input
    if not user_question.strip():
        return "Please type a question."

    # Step 1: Clean the user question
    cleaned_question = preprocess_text(user_question)

    # Step 2: Convert to TF-IDF vector
    user_vector = vectorizer.transform([cleaned_question])

    # Step 3: Calculate cosine similarity with all FAQ vectors
    similarity_scores = cosine_similarity(user_vector, faq_vectors)

    # Step 4: Find highest scoring FAQ
    best_match_index = similarity_scores.argmax()
    best_score = similarity_scores[0][best_match_index]

    # Step 5: Threshold check
    threshold = 0.2

    if best_score < threshold:
        return "I'm sorry, I couldn't find an answer to that question. Please contact the college office directly or try rephrasing your question."

    return df['answer'][best_match_index]

# Test it once
print(get_best_answer("how to get bonafide certificate"))
```

```python
# Testing with multiple questions to verify chatbot works correctly

test_questions = [
    "how to get bonafide certificate",
    "minimum attendance for exam",
    "scholarship application process",
    "library book borrowing rules",
    "what is the wifi password"  # this should trigger fallback
]

print("="*60)
for q in test_questions:
    print(f"\nQuestion: {q}")
    print(f"Answer: {get_best_answer(q)}")
    print("-"*60)
```

```python
# ============================================
# STEP 7: ANIMATED IMPRESSIVE EDUBOT UI
# Particles + Glow + Glassmorphism Design
# ============================================

import gradio as gr
import time

custom_css = """
/* Animated gradient background */
@keyframes gradientShift {
    0% { background-position: 0% 50%; }
    50% { background-position: 100% 50%; }
    100% { background-position: 0% 50%; }
}

@keyframes float {
    0%, 100% { transform: translateY(0px) rotate(0deg); opacity: 0.7; }
    33% { transform: translateY(-30px) rotate(120deg); opacity: 1; }
    66% { transform: translateY(-15px) rotate(240deg); opacity: 0.8; }
}

@keyframes pulse {
    0%, 100% { box-shadow: 0 0 10px #00d4ff, 0 0 20px #00d4ff, 0 0 40px #00d4ff; }
    50% { box-shadow: 0 0 20px #ff00ff, 0 0 40px #ff00ff, 0 0 80px #ff00ff; }
}

@keyframes slideIn {
    from { opacity: 0; transform: translateY(20px); }
    to { opacity: 1; transform: translateY(0px); }
}

@keyframes titleGlow {
    0%, 100% { text-shadow: 0 0 10px #00d4ff, 0 0 20px #00d4ff; }
    50% { text-shadow: 0 0 20px #ff00ff, 0 0 40px #ff00ff, 0 0 60px #ff00ff; }
}

@keyframes borderGlow {
    0%, 100% { border-color: #00d4ff; box-shadow: 0 0 10px #00d4ff; }
    50% { border-color: #ff00ff; box-shadow: 0 0 20px #ff00ff; }
}

@keyframes particle {
    0% { transform: translateY(100vh) rotate(0deg); opacity: 0; }
    10% { opacity: 1; }
    90% { opacity: 1; }
    100% { transform: translateY(-100vh) rotate(720deg); opacity: 0; }
}

/* Main container */
.gradio-container {
    background: linear-gradient(-45deg, #0a0a1a, #0d1b2a, #1a0a2e, #0a1628, #1a1a0a);
    background-size: 400% 400%;
    animation: gradientShift 8s ease infinite !important;
    font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif !important;
    min-height: 100vh !important;
    position: relative !important;
    overflow: hidden !important;
}

/* Floating particles */
.gradio-container::before {
    content: '🤖 ⭐ 💡 📚 🎓 ✨ 🔬 💫';
    position: fixed;
    top: 0;
    left: 0;
    width: 100%;
    height: 100%;
    font-size: 1.5em;
    animation: float 6s ease-in-out infinite;
    pointer-events: none;
    z-index: 0;
    opacity: 0.15;
    word-spacing: 100px;
    line-height: 200px;
}

/* Title */
.gradio-container h1 {
    color: #00d4ff !important;
    text-align: center !important;
    font-size: 2.5em !important;
    font-weight: 800 !important;
    animation: titleGlow 3s ease-in-out infinite !important;
    letter-spacing: 2px !important;
    padding: 20px 0 5px 0 !important;
    position: relative !important;
    z-index: 1 !important;
}

/* Subtitle */
.gradio-container .prose p {
    color: #a8c7fa !important;
    text-align: center !important;
    font-size: 1em !important;
    position: relative !important;
    z-index: 1 !important;
}

/* Glassmorphism card effect */
.gradio-container .block {
    background: rgba(255, 255, 255, 0.05) !important;
    backdrop-filter: blur(10px) !important;
    border: 1px solid rgba(0, 212, 255, 0.3) !important;
    border-radius: 20px !important;
    animation: borderGlow 4s ease-in-out infinite !important;
}

/* User message */
.message.user {
    background: linear-gradient(135deg,
        rgba(0, 212, 255, 0.2),
        rgba(0, 100, 200, 0.3)) !important;
    color: #ffffff !important;
    border-radius: 20px 20px 5px 20px !important;
    padding: 15px 20px !important;
    border: 1px solid rgba(0, 212, 255, 0.5) !important;
    box-shadow: 0 0 15px rgba(0, 212, 255, 0.3) !important;
    animation: slideIn 0.3s ease !important;
    backdrop-filter: blur(10px) !important;
}

/* Bot message */
.message.bot {
    background: linear-gradient(135deg,
        rgba(255, 0, 255, 0.1),
        rgba(100, 0, 200, 0.2)) !important;
    color: #e8f4fd !important;
    border-radius: 20px 20px 20px 5px !important;
    padding: 15px 20px !important;
    border: 1px solid rgba(255, 0, 255, 0.3) !important;
    box-shadow: 0 0 15px rgba(255, 0, 255, 0.2) !important;
    animation: slideIn 0.3s ease !important;
    backdrop-filter: blur(10px) !important;
}

/* Input box */
.gradio-container textarea {
    background: rgba(0, 10, 30, 0.8) !important;
    color: #00d4ff !important;
    border: 2px solid #00d4ff !important;
    border-radius: 15px !important;
    padding: 15px !important;
    font-size: 1em !important;
    backdrop-filter: blur(10px) !important;
    transition: all 0.3s ease !important;
}

.gradio-container textarea:focus {
    border-color: #ff00ff !important;
    box-shadow: 0 0 20px rgba(255, 0, 255, 0.5) !important;
    color: #ffffff !important;
}

/* Send button */
.gradio-container button.primary {
    background: linear-gradient(135deg, #00d4ff, #ff00ff) !important;
    color: #ffffff !important;
    border: none !important;
    border-radius: 15px !important;
    font-weight: 800 !important;
    font-size: 1em !important;
    padding: 15px 30px !important;
    cursor: pointer !important;
    transition: all 0.3s ease !important;
    box-shadow: 0 0 20px rgba(0, 212, 255, 0.5) !important;
    animation: pulse 3s ease-in-out infinite !important;
}

.gradio-container button.primary:hover {
    transform: translateY(-3px) scale(1.05) !important;
    box-shadow: 0 0 40px rgba(255, 0, 255, 0.8) !important;
}

/* Secondary buttons */
.gradio-container button.secondary {
    background: rgba(0, 212, 255, 0.1) !important;
    color: #00d4ff !important;
    border: 1px solid #00d4ff !important;
    border-radius: 12px !important;
    transition: all 0.3s ease !important;
}

.gradio-container button.secondary:hover {
    background: rgba(0, 212, 255, 0.3) !important;
    box-shadow: 0 0 15px rgba(0, 212, 255, 0.5) !important;
    transform: translateY(-2px) !important;
}

/* Chatbot container */
.gradio-container .chatbot {
    background: rgba(0, 5, 20, 0.7) !important;
    border: 1px solid rgba(0, 212, 255, 0.3) !important;
    border-radius: 20px !important;
    backdrop-filter: blur(15px) !important;
}

/* Stat cards */
.stat-card {
    text-align: center;
    background: rgba(0, 212, 255, 0.1);
    border: 1px solid rgba(0, 212, 255, 0.4);
    border-radius: 15px;
    padding: 12px;
    color: #a8c7fa;
    transition: all 0.3s ease;
    animation: borderGlow 4s ease-in-out infinite;
}

.stat-card:hover {
    background: rgba(0, 212, 255, 0.2);
    transform: translateY(-3px);
    box-shadow: 0 10px 30px rgba(0, 212, 255, 0.3);
}

/* Scrollbar */
::-webkit-scrollbar { width: 6px; }
::-webkit-scrollbar-track { background: rgba(0,0,0,0.3); }
::-webkit-scrollbar-thumb {
    background: linear-gradient(#00d4ff, #ff00ff);
    border-radius: 3px;
}

/* Example buttons */
.gradio-container .example-set button {
    background: rgba(0, 212, 255, 0.1) !important;
    color: #00d4ff !important;
    border: 1px solid rgba(0, 212, 255, 0.4) !important;
    border-radius: 25px !important;
    padding: 8px 18px !important;
    font-size: 0.85em !important;
    transition: all 0.3s ease !important;
    margin: 4px !important;
}

.gradio-container .example-set button:hover {
    background: linear-gradient(135deg, #00d4ff, #ff00ff) !important;
    color: #ffffff !important;
    transform: translateY(-2px) !important;
    box-shadow: 0 5px 20px rgba(0, 212, 255, 0.5) !important;
}
"""

# Category questions
CATEGORIES = {
    "📋 Admissions": "What documents are required for college admission?",
    "📝 Exams": "What is the attendance requirement to write exams?",
    "🎓 Certificates": "How do I apply for a bonafide certificate?",
    "💰 Scholarships": "What scholarships are available for students?",
    "💼 Placements": "How do I register for placement drives?",
    "🏠 Hostel": "How to apply for hostel accommodation?"
}

def get_confidence_bar(score):
    """Plain text confidence bar"""
    percentage = int(score * 100)
    filled = int(percentage / 5)
    empty = 20 - filled

    if percentage >= 70:
        label = "High Match 🟢"
        bar_char = "█"
    elif percentage >= 40:
        label = "Good Match 🟡"
        bar_char = "▓"
    else:
        label = "Low Match 🔴"
        bar_char = "░"

    bar = bar_char * filled + "░" * empty
    return f"\n\n📊 Confidence: [{bar}] {percentage}% — {label}"

def chatbot_response_ui(user_message, history):
    """Main response with confidence score"""
    if not user_message.strip():
        return "", history

    time.sleep(0.5)

    cleaned = preprocess_text(user_message)
    user_vector = vectorizer.transform([cleaned])
    scores = cosine_similarity(user_vector, faq_vectors)
    best_idx = scores.argmax()
    best_score = scores[0][best_idx]

    if best_score < 0.25:
        bot_reply = (
            "❌ Sorry! I could not find a relevant answer.\n\n"
            "This question is outside my current knowledge base.\n\n"
            "Please contact:\n"
            "📞 College Office: 044-XXXXXXXX\n"
            "📧 Email: info@college.edu\n"
            "🏢 Visit: Admin Block, Room 101"
            + get_confidence_bar(best_score)
        )
    else:
        answer = df["answer"][best_idx]
        matched_q = df["question"][best_idx]
        bot_reply = (
            f"✅ {answer}\n\n"
            f"📌 Matched with: \"{matched_q}\""
            + get_confidence_bar(best_score)
        )

    history.append((user_message, bot_reply))
    return "", history

def clear_chat():
    return [[None, welcome_msg]]

def retry_last(history):
    if len(history) > 1:
        last_msg = history[-1][0]
        history.pop()
        _, history = chatbot_response_ui(last_msg, history)
    return history

def ask_category(question, history):
    _, history = chatbot_response_ui(question, history)
    return history

# Welcome message
welcome_msg = """🤖 **Hello! I am EduBot!**

⚡ Your Smart College FAQ Assistant powered by AI!

I can instantly answer questions about:
🎓 Admissions & Documents
📝 Exams, Results & Revaluation
📜 Certificates & TC
💰 Scholarships & Fee Concession
💼 Placements & Internships
🏠 Hostel, Library & Transport
🏥 Medical & Grievance

**Type your question or click a category below! 👇**"""

# Build interface
with gr.Blocks(
    css=custom_css,
    title="EduBot — Smart College FAQ Assistant"
) as demo:

    # Title
    gr.Markdown("# 🎓 EduBot — Smart College FAQ Assistant")
    gr.Markdown(
        "#### ⚡ Powered by NLTK • TF-IDF Vectorization • Cosine Similarity Matching"
    )

    # Stat cards
    with gr.Row():
        gr.Markdown("""<div class='stat-card'>
        📚 <b style='color:#00d4ff'>40</b><br>
        <span style='font-size:0.85em'>FAQ Questions</span>
        </div>""")

        gr.Markdown("""<div class='stat-card'>
        🧠 <b style='color:#00d4ff'>TF-IDF</b><br>
        <span style='font-size:0.85em'>NLP Engine</span>
        </div>""")

        gr.Markdown("""<div class='stat-card'>
        🎯 <b style='color:#00d4ff'>Cosine</b><br>
        <span style='font-size:0.85em'>Similarity</span>
        </div>""")

        gr.Markdown("""<div class='stat-card'>
        ⚡ <b style='color:#00d4ff'>Instant</b><br>
        <span style='font-size:0.85em'>AI Response</span>
        </div>""")

    gr.Markdown("---")

    # Chat area
    chatbot_ui = gr.Chatbot(
        value=[[None, welcome_msg]],
        height=430,
        bubble_full_width=False,
        show_label=False,
        avatar_images=(
            "https://api.dicebear.com/7.x/avataaars/svg?seed=student",
            "https://api.dicebear.com/7.x/bottts/svg?seed=edubot&backgroundColor=0d1b2a"
        ),
        render_markdown=True
    )

    # Input
    with gr.Row():
        msg_input = gr.Textbox(
            placeholder="💬 Ask EduBot anything about your college...",
            show_label=False,
            scale=5,
            container=False
        )
        send_btn = gr.Button("Send ✨", variant="primary", scale=1)

    # Action buttons
    with gr.Row():
        clear_btn = gr.Button("🗑️ Clear Chat", variant="secondary")
        retry_btn = gr.Button("🔄 Retry", variant="secondary")

    gr.Markdown("---")

    # Category buttons
    gr.Markdown(
        "<div style='color:#00d4ff;font-weight:700;"
        "font-size:1em;margin-bottom:8px'>"
        "⚡ Quick Categories — Click to get instant answers:</div>"
    )

    with gr.Row():
        cat_btns = []
        for label, question in CATEGORIES.items():
            btn = gr.Button(label, variant="secondary")
            cat_btns.append((btn, question))

    # Examples
    gr.Markdown(
        "<div style='color:#a8c7fa;"
        "font-size:0.85em;margin-top:10px'>"
        "💡 Or try these example questions:</div>"
    )

    gr.Examples(
        examples=[
            "How to apply for bonafide certificate?",
            "What is minimum attendance for exams?",
            "How to register for campus placement?",
            "What scholarships are available?",
            "How to get transfer certificate?",
            "How to pay college fees?"
        ],
        inputs=msg_input
    )

    # Footer
    gr.Markdown("""---
<div style='text-align:center;color:#4a6fa5;
            font-size:0.8em;padding:10px'>
    🎓 EduBot | Built with ❤️ using
    NLTK • TF-IDF • Cosine Similarity • Gradio<br>
    🤖 Retrieval Based NLP Chatbot |
    40 FAQs | College Helpdesk Assistant
</div>""")

    # Connect everything
    send_btn.click(
        chatbot_response_ui,
        inputs=[msg_input, chatbot_ui],
        outputs=[msg_input, chatbot_ui]
    )

    msg_input.submit(
        chatbot_response_ui,
        inputs=[msg_input, chatbot_ui],
        outputs=[msg_input, chatbot_ui]
    )

    clear_btn.click(clear_chat, outputs=[chatbot_ui])
    retry_btn.click(retry_last, inputs=[chatbot_ui], outputs=[chatbot_ui])

    for btn, question in cat_btns:
        btn.click(
            fn=lambda history, q=question: ask_category(q, history),
            inputs=[chatbot_ui],
            outputs=[chatbot_ui]
        )

# Launch
demo.launch(debug=True, share=True)
```
